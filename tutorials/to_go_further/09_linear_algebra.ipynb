## Definition of vectors

A vector is a mathematical object that has both magnitude and direction. It can be represented as a line segment with a starting point and an end point. The magnitude of a vector is its length, and the direction of a vector is the angle that it makes with a reference direction.

They can be used to represent a wide variety of physical quantities, such as force, velocity, and acceleration (well known from your first lectures in mechanical physics). To get a quick refresh of vector spaces, one can visit this great [Wikipedia page](https://en.wikipedia.org/wiki/Vector_space).

A 1-element vector is, for example, a point in space. A 2-element vector is an arrow or even a point on an $x-y$ plane. 3-element vectors represent the same thing as the previous two, but in 3D space. N-dimensional vectors is something very hard to imagine... But you will find some of them in this course.

### What is their purpose in programming or in data science?

They introduce the notion of observations and predictions -- which is really useful in power system analysis.
For example, one can imagine a power grid as a network of interconnected wires and power plants. The voltage and current at any point in the grid can be represented by a vector. The magnitude of the vector represents the voltage, and the direction of the vector represents the current.

The following vector could represent the voltage and current at a particular point in the grid:

**grid_point** $= \begin{pmatrix}
10.5 \\
5.2 \\
3.25 \\
7.0
\end{pmatrix}$

This means that the voltage at this point is $120~\text{V}$, and the current is $10~\text{A}$.

In NumPy a vector is a one-dimensional array of numbers, and we can define it in Python as follows:

````python
[10.5, 5.2, 3.25, 7.0]
````

## Plotting vectors

````python
import matplotlib.pyplot as plt
````

### 2D vectors

````python
u = np.array([2, 5])
v = np.array([3, 1])

x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=["r", "b"])
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
````

Vectors can also be represented as arrows. We can plot them as follows:

````python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
````

````python
plot_vector2d(u, color="r")
plot_vector2d(v, color="b")
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
````

### 3D vectors

````python
a = np.array([1, 2, 8])
b = np.array([5, 6, 3])
````

````python
subplot3d = plt.subplot(111, projection='3d')
x_coords, y_coords, z_coords = zip(a,b)
subplot3d.scatter(x_coords, y_coords, z_coords)
subplot3d.set_zlim3d([0, 9])
plt.show()
````

````python
def plot_vectors3d(ax, vectors3d, z0, **options):
    for v in vectors3d:
        x, y, z = v
        ax.plot([x,x], [y,y], [z0, z], color="gray", linestyle='dotted', marker=".")
    x_coords, y_coords, z_coords = zip(*vectors3d)
    ax.scatter(x_coords, y_coords, z_coords, **options)

subplot3d = plt.subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vectors3d(subplot3d, [a,b], 0, color=("r","b"))
plt.show()
````
